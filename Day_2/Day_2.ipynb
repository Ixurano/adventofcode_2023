## Part one: Find which game would have been possible
Determine which games would have been possible if the bag had been loaded with only 12 red cubes, 13 green cubes, and 14 blue cubes. What is the sum of the IDs of those games?

```python
import csv
import regex
file_path = 'input.csv'

game_data = {'Name': '', 'Red': 0, 'Green': 0, 'Blue': 0}
total = 0
with open(file_path, 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=";")
    for row in reader:
        game_header = row[0].split(': ')
        game_name = game_header[0]
        game_name_number = game_name.split('Game ')
        initial_set = game_header[1]
        game_data['Name'] = game_name_number[1]
        row.append(initial_set)
        for set_data in row[1:]:
            set_values = set_data.split(',')
            if len(set_values) != 0: 
                for value in set_values:
                    match:str = regex.findall(r'(\d+)\s*([a-zA-Z]+)', value)
                    quantity, color = match[0]
                    if color.lower() == 'red':
                        if int(quantity) > 12:
                            game_data['Name'] = 'invalid'
                        else:  
                            game_data['Red'] = int(quantity)
                    elif color.lower() == 'green':
                        if int(quantity) > 13:
                            game_data['Name'] = 'invalid'
                        else:
                            game_data['Green'] = int(quantity)
                    elif color.lower() == 'blue':
                        if int(quantity) > 14:
                            game_data['Name'] = 'invalid'
                        else:
                            game_data['Blue'] = int(quantity)

        #print(game_data)
        if(game_data['Name'] != 'invalid'):
            total += int(game_data['Name'])
        #put total ++ here 
        # Reset the counters for the next iteration
        game_data = {'Name': '', 'Red': 0, 'Green': 0, 'Blue': 0}
        print(total)
        
```
